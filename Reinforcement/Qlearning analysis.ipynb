```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open('history.pckl', 'rb')
history = pickle.load(f)
f.close()
```

```python
plt.figure(figsize=(10,8))

plt.plot((history["loss"]))
```

```python
plt.plot(history["qval"])
```

```python
plt.figure(figsize=(10,8))

print(np.mean(history["score"][1500:]))
plt.plot(history["score"])
plt.tick_params(labelsize=16)
plt.grid()
plt.title("Training Score Per Game", fontsize=24)
plt.xlabel("Game", fontsize=20)
plt.ylabel("Score", fontsize=20)
```

```python
print(len(history["moves"]))
plt.plot(history["moves"])
```

```python
plt.figure(figsize=(10,8))

plt.plot(history["acc"])
plt.tick_params(labelsize=16)

plt.grid()
plt.title("Training Accuracy Per Game Q-Learning", fontsize=24)
plt.xlabel("Game", fontsize=20)
plt.ylabel("Accuracy", fontsize=20)
```

```python
mean = []
std = []
for i in range(0, len(history["acc"]), 50):
    mean.append(np.mean(history["acc"][i:i+50]))
    std.append(np.std(history["acc"][i:i+50]))

```

```python
plt.figure(figsize=(10,8))

plt.plot(mean)
plt.tick_params(labelsize=16)
plt.fill_between(list(range(len(mean))), np.array(mean)-np.array(std), np.array(mean)+np.array(std), color=(1,0,0,0.3))


plt.grid()
plt.title("Accuracy Mean per 50 games windows", fontsize=24)
plt.xlabel("Window Number", fontsize=20)
plt.ylabel("Mean Accuracy", fontsize=20)
```

```python
meanScore = []
stdScore = []
for i in range(0, len(history["score"]), 50):
    meanScore.append(np.mean(history["score"][i:i+50]))
    stdScore.append(np.std(history["score"][i:i+50]))
```

```python
plt.figure(figsize=(10,8))

plt.plot(meanScore)
plt.fill_between(list(range(len(meanScore))), np.array(meanScore)-np.array(stdScore), np.array(meanScore)+np.array(stdScore), color=(1,0,0,0.3))
plt.tick_params(labelsize=16)

plt.grid()
plt.title("Score Mean per 50 games windows", fontsize=24)
plt.xlabel("Window Number", fontsize=20)
plt.ylabel("Mean Score", fontsize=20)
```
