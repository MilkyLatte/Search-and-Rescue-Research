```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = open('./test/SaltPepperHistory1000.pckl', 'rb')
history = pickle.load(f)
f.close()
```

```python
plt.figure(figsize=(10,8))
plt.plot(history['accuracy'])
plt.tick_params(labelsize=16)

plt.grid()
plt.title("Accuracy on the validation set Supervised Learning", fontsize=24)
plt.xlabel("Number of batches * 100", fontsize=20)
plt.ylabel("Accuracy", fontsize=20)
```

```python
plt.figure(figsize=(10,8))
plt.tick_params(labelsize=16)

plt.plot(history['loss'])
plt.grid()
plt.title("Loss on the validation set Supervised Learning", fontsize=24)
plt.xlabel("Number of batches * 100", fontsize=20)
plt.ylabel("Loss", fontsize=20)
```

```python
mean = []
std = []
for i in range(0, len(history["accuracy"]), 10):
    mean.append(np.mean(history["accuracy"][i:i+10]))
    std.append(np.std(history["accuracy"][i:i+10]))

plt.plot(mean)
plt.plot(std)
```

```python
plt.figure(figsize=(10,8))

plt.plot(mean)
plt.tick_params(labelsize=16)
plt.fill_between(list(range(len(mean))), np.array(mean)-np.array(std), np.array(mean)+np.array(std), color=(1,0,0,0.3))


plt.grid()
plt.title("Accuracy Mean per 50 games windows", fontsize=24)
plt.xlabel("Window Number", fontsize=20)
plt.ylabel("Mean Accuracy", fontsize=20)
```
